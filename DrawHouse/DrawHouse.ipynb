```python
import turtle

# Tạo một đối tượng Turtle
ngoi_nha = turtle.Turtle()
cay = turtle.Turtle()
mat_troi = turtle.Turtle()
# Vẽ ngôi nhà
ngoi_nha.penup()
ngoi_nha.goto(-100, -100)
ngoi_nha.pendown()
ngoi_nha.begin_fill()
ngoi_nha.color("orange")  # Màu cam cho ngôi nhà
for _ in range(4):
    ngoi_nha.forward(200)
    ngoi_nha.left(90)
ngoi_nha.end_fill()

# Vẽ mái nhà
ngoi_nha.penup()
ngoi_nha.goto(-100, 100)
ngoi_nha.pendown()
ngoi_nha.begin_fill()
ngoi_nha.color("brown")  # Màu nâu cho mái nhà
ngoi_nha.goto(0, 200)
ngoi_nha.goto(100, 100)
ngoi_nha.goto(-100, 100)
ngoi_nha.end_fill()

# Vẽ cửa
ngoi_nha.penup()
ngoi_nha.goto(-30, -100)
ngoi_nha.pendown()
ngoi_nha.begin_fill()
ngoi_nha.color("blue")  # Màu xanh cho cửa
for _ in range(4):
    ngoi_nha.forward(60)
    ngoi_nha.left(90)
ngoi_nha.end_fill()

# Vẽ cửa sổ
ngoi_nha.penup()
ngoi_nha.goto(50, -20)
ngoi_nha.pendown()
ngoi_nha.begin_fill()
ngoi_nha.color("yellow")  # Màu vàng cho cửa sổ
for _ in range(4):
    ngoi_nha.forward(40)
    ngoi_nha.left(90)
ngoi_nha.end_fill()

# Vẽ thân cây
cay.penup()
cay.goto(-270, -100)
cay.pendown()
cay.begin_fill()
cay.color("brown")  # Màu nâu cho thân cây
cay.goto(-230, -100)
cay.goto(-230, -30)
cay.goto(-270, -30)
cay.goto(-270, -100)
cay.end_fill()

# Vẽ lá cây
cay.penup()
cay.goto(-310, -50)
cay.pendown()
cay.begin_fill()
cay.color("green")  # Màu xanh cho lá cây
cay.goto(-250, 50)
cay.goto(-190, -50)
cay.goto(-250, -20)
cay.goto(-310, -50)
cay.end_fill()

# Vẽ mặt trời
mat_troi.penup()
mat_troi.goto(200, 200)
mat_troi.pendown()
mat_troi.begin_fill()
mat_troi.color("yellow")  # Màu vàng cho mặt trời

# Vẽ mặt trời bằng một đường tròn lớn
mat_troi.circle(30)

mat_troi.end_fill()

# Vẽ tia nắng
mat_troi.penup()
mat_troi.goto(200, 230)
mat_troi.pendown()
mat_troi.color("yellow")  # Màu vàng cho tia nắng

for _ in range(12):
    mat_troi.goto(200, 230)
    mat_troi.forward(40)
    mat_troi.backward(40)
    mat_troi.left(30)


# Đóng cửa sổ khi kết thúc
turtle.done()
```
